```python
# chargement du dataset dans le dataframe
import pandas as pd

df = pd.read_csv('/kaggle/input/customer-support-on-twitter/sample.csv')
df.head() # affichage des 1ers lignes du dataframe
```

```python
# lower text field in dataframe
df['text_cleaned'] = df['text'].str.lower()
df.head()
```

```python
# remove punctuation from text field
import string

df['text_cleaned'] = df['text_cleaned'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation)))
df.head()
```

```python
# remove stopwords
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df.head()
```

```python
# Identifying most and least frequent words and removing them
from nltk.probability import FreqDist # frequent distrabution

fdist = FreqDist(' '.join(df['text_cleaned']).split()) # each word is mapped to its frequency inside the text field
common_words = fdist.most_common(10)  # 10 most frequent
rare_words = fdist.most_common()[-10:]  # 10 least frequent

common_words = [word[0] for word in common_words] # extracting only the words not the frequencies
rare_words = [word[0] for word in rare_words] # ...

# removing them and join what's left in the text
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([word for word in x.split() if word not in common_words + rare_words]))
df.head()
```

```python
# Stemming & Lemmatisation
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))
df.head()
#df['text_cleaned'] = df['text_cleaned'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
#df.head()
```

```python
# Regular expressions library
import re

# Remove emojis
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: re.sub(r':\)|;\)|:-\)|\(-:|:\'\(|:-\(|:\)|:\(|:D|xD|XD|:P|:p', '', x))

# Remove emoticons
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: re.sub(r'[^\x00-\x7F]+', '', x))

# Removing URLs
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: re.sub(r'http\S+', '', x))

# Removing html tags
df['text_cleaned'] = df['text_cleaned'].apply(lambda x: re.sub(r'<.*?>', '', x))

df.head()
```
